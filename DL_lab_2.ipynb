```python
import random

def perceptron(x, wt):
    s = 0
    for i in range(len(x)):
        s += x[i] * wt[i]
    return 1 if s >= 0 else 0


x = [1, 1]

wt = [random.randint(-1, 1) for _ in range(len(x))]
print("Random Weights are", wt)

# Output
output = perceptron(x, wt)
print("Output:", output)
```
