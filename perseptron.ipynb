```python
def perceptron(inputs):
    # Manually defined weights and bias
    weights = [1, 1]   # w1, w2
    bias = -1.5

    # Step activation function
    weighted_sum = 0
    for i in range(len(inputs)):
        weighted_sum += inputs[i] * weights[i]

    weighted_sum += bias

    if weighted_sum >= 0:
        return 1
    else:
        return 0


# Test the perceptron (AND gate)
test_inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

for x in test_inputs:
    print(f"Input: {x}, Output: {perceptron(x)}")
```

```python
def perceptron(x):
    w1 = 1
    w2 = 2
    b = -1.5

    s = x[0]*w1 + x[1]*w2 + b

    if s >= 0:
        return 1
    else:
        return 0


# Example inputs
inputs = [[0,0], [0,1], [1,0], [1,1]]

for i in inputs:
    print(i, perceptron(i))
```
